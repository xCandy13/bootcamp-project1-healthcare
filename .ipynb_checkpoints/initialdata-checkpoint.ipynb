```python
#import dependencies
import requests
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from config import covid_api
```

```python
#setup base url and perform initial get request
covid_base_url = 'https://api.covidactnow.org/v2/states.json?apiKey='+covid_api
covid_json = requests.get(covid_base_url).json()
#print out initial data
print(json.dumps(covid_json, indent=4))
```

```python
#look at csv file
covid_csv = requests.get('https://api.covidactnow.org/v2/states.csv?apiKey='+covid_api)
print(covid_csv.csv())
```
